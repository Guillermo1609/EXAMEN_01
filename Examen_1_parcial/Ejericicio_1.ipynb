```
import math 

print("CALCULADORA DE RAICES PARA ECUACIONES TIPO\n            **ax^2+bx+c=0**\n")
print("-->Ingresa los valores de a, b, c:")
a=float (input("*Valor para a:"))
b=float (input("*Valor para b:"))
c=float (input("*Valor para c:"))
print("\nLa ecuación ingresada es: ",a,"x^2+",b,"x+",c,"=0")
disc=(pow(b,2)-4*a*c)
print("El discriminante (b^2+4ac) es:",disc)
while disc<0:    #disc=discriminante
  print("\nERROR:El discriminante debe de ser positivo.\n""Vuelva a ingresar los valores para los coeficientes de a, b, c:")
  a=float (input("\nValor para a:"))
  b=float (input("Valor para b:"))
  c=float (input("Valor para c:"))
  print("\nLa ecuación ingresada es:",a,"x^2+",b,"x+",c,"=0")
  disc=(pow(b,2)-4*a*c)
  print("El discriminante (b^2+4ac) es: ",disc)
x1=(-b+pow(disc,1/2))/(2*a)
print("\n\nEl resultado de x1 es:",round(x1,10))
x2=(-b-pow(disc,1/2))/(2*a)
print("El resultado de x2 es:",round(x2,10))
print("\n              Fin del programa")
```
