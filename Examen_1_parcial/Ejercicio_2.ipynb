```
import math

def funcion(x):
  e=2.7182818284
  func=pow(e,-x)
  return func
  
print("f(x)=e^-x")
val=[]
y=[]
contador=-4
while (-4<=contador<10):
  x=contador
  val.append(funcion(x))
  y.append(x)
  contador+=1
  if funcion(x)<0.0025:
    break
print("Los valores de la función evaluada de -4 a 10 son:\n",y,"\n",val)
```
